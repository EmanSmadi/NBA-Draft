```python
# making a function 
def makedataframe(rows):
    list_rows = []
    for row in rows:
        global df, df1, df2
        cells = row.find_all('td')
        str_cells = str(cells)
        clean = re.compile('<.*?>')
        clean2 = (re.sub(clean, '',str_cells))
        list_rows.append(clean2)
        df = pd.DataFrame(list_rows)
        df1 = df[0].str.replace(", Jr.,"," Jr.,",regex=True)
        df2 = df1.str.split(',', expand=True)
        df2[0] = df2[0].str.strip('[')
        #df2[22] = df2[22].str.strip(']')
    return(df2)
```

```python
## add all the data into one table
master_df = pd.DataFrame()
for url in urls_Advanced:
    html = urlopen(url)
    soup = BeautifulSoup(html, 'lxml')
    rows = soup.find_all('tr')
    # creating a data frame for each url 
    mydf = makedataframe(rows)
    
    # add identifiers to dataframe from each url 
    mydf['year'] = url[41:45] 
    mydf['position'] = url[81:83]
    mydf['class'] = url[71:73]
    
    # adding data frame to the master data frame 
    master_df = pd.concat([master_df,mydf])
    
master_df.head(200)
```
