```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
from ipykernel import kernelapp as app
import warnings
warnings.filterwarnings("ignore")
import re
```

```python
years = np.arange(2009,2019)
print(years)
```

```python
draft_results = list()
# Generate urls
for year in years:
    url = f'https://basketball.realgm.com/nba/draft/past_drafts/{year}/'
    draft_results.append(url)
```

```python
# making a function 
def makedataframe(rows):
    list_rows = []
    for row in rows:
        global df, df1, df2
        cells = row.find_all('td')
        str_cells = str(cells)
        clean = re.compile('<.*?>')
        clean2 = (re.sub(clean, '',str_cells))
        list_rows.append(clean2)
        df = pd.DataFrame(list_rows)
        df1 = df[0].str.replace(", Jr.,"," Jr.,",regex=True)
        df2 = df1.str.split(',', expand=True)
        df2[0] = df2[0].str.strip('[')
        #df2[22] = df2[22].str.strip(']')
    return(df2)
```

```python
## adding all results into 1 table
master_df = pd.DataFrame()
for url in draft_results:
    html = urlopen(url)
    soup = BeautifulSoup(html, 'lxml')
    rows = soup.find_all('tr')
    # creating a data frame for each url 
    mydf = makedataframe(rows)
    
    # add identifiers to dataframe from each url 
    mydf['year'] = url[52:56]
    # adding data frame to the master data frame 
    master_df = pd.concat([master_df,mydf])
```
